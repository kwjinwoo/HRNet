```python
from OXFORD import OXFORDLoader
from PASCALContext import PASCALContextLoader
from PASCAL2012 import PASCAL2012Loader
import matplotlib.pyplot as plt
import numpy as np
```

```python
loader = PASCALContextLoader(train_path='./tfrecords/train_PASCALContext.tfrecord',
                                     val_path='./tfrecords/val_PASCALContext.tfrecord',
                                     height=480, width=480, batch_size=4)
train_ds, val_ds = loader.get_train_val_ds()
# loader = PASCAL2012Loader(train_path='./tfrecords/train_PASCAL2012.tfrecord',
#                                      val_path='./tfrecords/val_PASCAL2012.tfrecord',
#                                      height=480, width=480, batch_size=4)
# train_ds, val_ds = loader.get_train_val_ds()
# loader = OXFORDLoader(train_path='./tfrecords/train_OXFORD.tfrecord',
#                                      val_path='./tfrecords/val_OXFORD.tfrecord',
#                                      height=480, width=480, batch_size=4)
# train_ds, val_ds = loader.get_train_val_ds()
```

```python
for i in train_ds.take(1):
    sample = i
    
sample[0].shape, sample[1].shape
```

```python
plt.figure(figsize=(16, 16))
for i in range(4):
    mask = sample[1][i].numpy()[..., 0]
    # mask = np.where(mask == 21, 255, mask)
    plt.subplot(2, 2, i + 1)
    plt.title(np.unique(mask))
    plt.imshow(mask)
```

```python
plt.figure(figsize=(16, 16))
for i in range(4):
    image = sample[0][i].numpy()
    plt.subplot(2, 2, i + 1)
    plt.imshow(image)
```
