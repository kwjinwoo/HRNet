```python
import data_utils as du
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_ds = du.get_pascal_context_dataset('./train_PASCAL_context_480_480.tfrecord', 'train', 480, 480, 4)
val_ds = du.get_pascal_context_dataset('./val_PASCAL_context_480_480.tfrecord', 'val', 480, 480, 4)
```

```python
for i in train_ds.take(2):
    sample = i
    
sample[0].shape, sample[1].shape
```

```python
plt.figure(figsize=(16, 16))
for i in range(4):
    mask = sample[1][i].numpy()[..., 0]
    plt.subplot(2, 2, i + 1)
    plt.title(np.unique(mask))
    plt.imshow(mask)
```

```python
plt.figure(figsize=(16, 16))
for i in range(4):
    image = sample[0][i].numpy()
    plt.subplot(2, 2, i + 1)
    plt.imshow(image)
```
