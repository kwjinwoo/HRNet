```python
from PASCAL_2012 import PASCAL2012
import matplotlib.pyplot as plt
import numpy as np
```

```python
pascal = PASCAL2012()
train_ds = pascal.get_pascal2012_dataset('./train_PASCAL2012.tfrecord', 'train', 480, 480, 4)
val_ds = pascal.get_pascal2012_dataset('./val_PASCAL2012.tfrecord', 'val', 480, 480, 4)
```

```python
for i in train_ds.take(1):
    sample = i
    
sample[0].shape, sample[1].shape
```

```python
plt.figure(figsize=(16, 16))
for i in range(4):
    mask = sample[1][i].numpy()[..., 0]
    mask = np.where(mask == 21, 255, mask)
    plt.subplot(2, 2, i + 1)
    plt.title(np.unique(mask))
    plt.imshow(mask)
```

```python
plt.figure(figsize=(16, 16))
for i in range(4):
    image = sample[0][i].numpy()
    plt.subplot(2, 2, i + 1)
    plt.imshow(image)
```
